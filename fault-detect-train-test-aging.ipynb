```python
import numpy 
Vald_X = numpy.load("./data_ind_99_clean_nor_vald.npz")["x"]
Vald_Y = numpy.load("./data_ind_99_clean_nor_vald.npz")["y"]
```

```python
import keras
from keras import layers
from keras.optimizers import SGD
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
import numpy as np
import pandas as pd
from IPython.display import display


Train_X = np.load("./data_pract.npz")["x"]
Train_Y = np.load("./data_pract.npz")["y"]

Train_X = Train_X[coef_X == Ind_90]
Train_Y = Train_Y[coef_X == Ind_90]




inputs = keras.Input(shape = (Train_X.shape[1], Train_X.shape[2], Train_X.shape[3]))

x = layers.Conv2D(filters = 64, kernel_size = (3, 3),strides=(2, 2), activation = "relu") (inputs)
x = layers.BatchNormalization()(x) 
x = layers.MaxPooling2D(pool_size=(1, 1))(x)

x = layers.Flatten()(x)

x = layers.Dense(256, activation="relu")(x) 

outputs = layers.Dense(Train_Y.shape[1], activation='softmax')(x)
model = keras.Model(inputs, outputs, name="model") 
optimizer = SGD(learning_rate=0.001)  
model.compile(optimizer=optimizer, loss='CategoricalCrossentropy', metrics=['accuracy']) 
model.summary()
# Duplicate the training data
Train_X_duplicated = np.repeat(Train_X, 4, axis=0)
Train_Y_duplicated = np.repeat(Train_Y, 4, axis=0)

history = model.fit(Train_X_duplicated, Train_Y_duplicated, epochs=100,batch_size = 64, verbose=2, validation_data=(Vald_X, Vald_Y))

print ('Training   ', model.metrics_names,' = ',model.evaluate(Train_X, Train_Y, verbose=0))
print ('Validation ', model.metrics_names,' = ',model.evaluate(Vald_X, Vald_Y, verbose=0))



show_acc_los(history)



Predict_test_Y = model.predict(Train_X)

y_true = np.argmax(Train_Y, axis=1)
y_pred = np.argmax(Predict_test_Y, axis=1)
# assume y_true and y_pred are your ground truth and predicted labels
cm = confusion_matrix(y_true, y_pred)

# normalize the confusion matrix by row
cm_norm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
df_cm = pd.DataFrame(np.round(cm_norm, 2))
display(df_cm)
# create a list of label names
labels = ['class 0', 'class 1', 'class 2', 'class 3', 'class 4', 'class 5', 'class 6', 'class 7', 'class 8']

# create a heatmap of the confusion matrix
# plot the heatmap
fig, ax = plt.subplots(figsize=(10, 8))
sns.heatmap(cm_norm, annot=True, ax=ax, xticklabels=labels, yticklabels=labels, cmap= 'Blues')

# add labels and title
plt.ylabel('True Label')
plt.xlabel('Predicted Label')
plt.title('Confusion Matrix')

plt.show()



y_true = np.argmax(Vald_Y, axis=1)
y_pred = np.argmax(Predict_test_Y, axis=1)
# assume y_true and y_pred are your ground truth and predicted labels
cm = confusion_matrix(y_true, y_pred)

# normalize the confusion matrix by row
cm_norm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
df_cm = pd.DataFrame(np.round(cm_norm, 2))
display(df_cm)
# create a list of label names
labels = ['class 0', 'class 1', 'class 2', 'class 3', 'class 4', 'class 5', 'class 6', 'class 7', 'class 8']

# create a heatmap of the confusion matrix
# plot the heatmap
fig, ax = plt.subplots(figsize=(10, 8))
sns.heatmap(cm_norm, annot=True, ax=ax, xticklabels=labels, yticklabels=labels, cmap= 'Blues')

# add labels and title
plt.ylabel('True Label')
plt.xlabel('Predicted Label')
plt.title('Confusion Matrix')

plt.show()
```
